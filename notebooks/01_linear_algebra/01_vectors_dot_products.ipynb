# Linear Algebra 01: Vectors and Dot Products

## Learning Goals

- Understand a vector as a list of coordinates.
- Interpret 2D vectors as arrows from the origin.
- Compute dot products.
- Interpret dot products as alignment, opposition, or orthogonality.
- Connect dot products to signal pattern matching.

## Concept

A vector like `[2, 3]` can be read as `x = 2`, `y = 3`. In geometry, that means an arrow from `(0, 0)` to `(2, 3)`.

A dot product multiplies matching entries and adds the results:

`[a, b] . [c, d] = a*c + b*d`

Interpretation:

- Positive: same-ish direction
- Zero: perpendicular / orthogonal
- Negative: opposite-ish direction

## Worked Example

`[1, 2, 3] . [4, 0, -1] = 1*4 + 2*0 + 3*(-1) = 1`

The result is positive, so the vectors have some alignment.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 0, -1])

np.dot(a, b)
```
