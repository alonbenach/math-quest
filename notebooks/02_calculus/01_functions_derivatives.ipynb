## Learning Goals

- Review functions as input-output rules.
- Interpret derivatives as rates of change.
- Compute basic power-rule derivatives.
- Connect derivatives to changing signals.

## Concept

A function takes an input and returns an output. A derivative tells us how fast the output changes as the input changes.

For powers of `x`, the power rule says:

`d/dx x^n = n*x^(n-1)`

## Worked Example

`d/dx (x^3 + 2x) = 3x^2 + 2`

```python
import sympy as sp

x = sp.symbols('x')
sp.diff(x**3 + 2*x, x)
```
