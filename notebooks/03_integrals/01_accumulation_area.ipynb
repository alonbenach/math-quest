## Concept

An integral can be interpreted as adding up infinitely many tiny pieces. Geometrically, a definite integral often represents area under a curve.

For powers of `x`, the basic rule is:

`integral x^n dx = x^(n+1)/(n+1) + C`, as long as `n != -1`.

## Worked Example

`integral 2x dx = x^2 + C`

We can check this because the derivative of `x^2 + C` is `2x`.

```python
import sympy as sp

x = sp.symbols('x')
sp.integrate(2*x, x)
```
