## Learning Goals

## Concept

A matrix-vector product takes a vector as input and returns another vector as output. Geometrically, a matrix can rotate, stretch, squash, shear, or reflect vectors.

## Worked Example

```text
[ 2  0 ] [3] = [ 6]
[ 1 -1 ] [4]   [-1]
```

The first output entry is `2*3 + 0*4 = 6`.

The second output entry is `1*3 + (-1)*4 = -1`.

````python
import numpy as np

A = np.array([[2, 0], [1, -1]])
v = np.array([3, 4])

A @ v
````
