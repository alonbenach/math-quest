## Concept

A point like `(2, 3)` names a location. A vector like `[2, 3]` can describe a movement: move 2 units in the x direction and 3 units in the y direction.

## Worked Example

The line with slope `2` passing through `(0, 3)` is:

`y = 2x + 3`

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)
y = 2*x + 3

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.plot(x, y)
plt.grid(True)
plt.show()
```
