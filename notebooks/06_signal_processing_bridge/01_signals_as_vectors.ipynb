# Signal Processing Bridge 01: Signals as Vectors

## Learning Goals

- Treat sampled signals as vectors.
- Use dot products to compare signal patterns.
- Recognize orthogonal signals as non-overlapping patterns.
- Prepare for Fourier ideas.

## Concept

A sampled signal like `[1, 0, -1, 0]` can be treated as a vector. Each entry is the signal value at a time sample.

Dot products compare a signal against a pattern.

## Worked Example

`[1, 0, -1, 0] . [1, 0, -1, 0] = 2`

The positive result means the signal matches the pattern.

`[1, 0, -1, 0] . [0, 1, 0, -1] = 0`

The zero result means the patterns are orthogonal.

```python
import numpy as np

signal = np.array([1, 0, -1, 0])
matching_pattern = np.array([1, 0, -1, 0])
orthogonal_pattern = np.array([0, 1, 0, -1])

np.dot(signal, matching_pattern), np.dot(signal, orthogonal_pattern)
```
